```python
import pandas as pd
from matplotlib import pyplot as plt


df = pd.read_csv("pokedex.csv")
df
```

```python
data = pd.DataFrame(df,
                        columns=['HP','Attack','Defense'])
data
```

```python
#hp = data["HP"]
attack= data["Attack"]
defense= data["Defense"]
#total = data["Total"]

#fig, (hp,attack,defense,total) = plt.subplots(nrow=4, ncols=1, sharey=True)


fig, ax = plt.subplots()

# grap to show the density of powerful pokemon based on attack and defense
scatter = ax.scatter(attack, defense, color='blue')
      
ax.set(xlabel='Attack',
       ylabel='Defense',
       title='Density of Powerful Pokemon')

plt.show()
```
