```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv("annualcheeseconsumed.txt")

year =  data["Year"]
cheddar = data["Cheddar"]
swiss = data["Swiss"]

fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, sharex =True)

ax1.plot(year, cheddar, label="Avg. Cheddar Consumed")
ax2.plot(year, swiss, label="Avg. Swiss Consumed")

ax1.set_title('American.....')
ax1.set_xlabel('Year')
ax1.set_ylabel('Chesse')
ax1.legend()

ax2.set_xlabel('Year')
ax2.set_ylabel('Chesse')
ax1.legend()
```
